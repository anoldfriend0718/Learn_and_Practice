## Batch Processing

```python
%%time
data_x = []
data_y = []
print('Image folders are:')
for i in range(len(sedimentary_class)):
    pic_folder=image_folder+sedimentary_class[i]+'\\'
    coll = fnmatch.filter(os.listdir(pic_folder), img_format)
    print(pic_folder,'Number of images:',len(coll))
    t = 0
    for j in range(len(coll)):
        img = batch_process(pic_folder+coll[j],x_shape,y_shape)
        data_x.append(img)
        data_y.append(sedimentary_class[i])
data_X = np.array(data_x)
data_Y = np.array(data_y)
print('The shape of feature set (set of image arrays) is: ',data_X.shape)
print('The shape of label is: ',data_Y.shape)
```
