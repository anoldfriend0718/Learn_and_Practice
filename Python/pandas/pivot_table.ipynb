```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
titanic=sns.load_dataset("titanic")
```

```python
titanic.head()
```

```python
titanic.loc[titanic.isnull()["age"]].head()
```

```python
titanic.groupby("sex")["survived"].mean()
```

```python
titanic.groupby(["sex","class"])["survived"].mean().unstack()
```

```python
titanic.pivot_table(values="survived",index=["sex"],columns="class",aggfunc=np.mean)
```

```python
age=pd.cut(titanic["age"],[0,18,80])
age.head()
```

```python
titanic.pivot_table(values="survived",index=["sex",age],columns="class")
```

```python
titanic.pivot_table(index="sex",columns="class",aggfunc={"survived":np.sum,"fare":np.mean})
```
