```python
import pandas as pd
import numpy as np
```

```python
pop=pd.read_csv("data-USstates/state-population.csv")
area=pd.read_csv("data-USstates/state-areas.csv")
abbrevs=pd.read_csv("data-USstates/state-abbrevs.csv")
```

```python
pop.head()
```

```python
area.head()
```

```python
abbrevs.head()
```

```python
merged=pd.merge(pop,abbrevs,left_on="state/region",right_on="abbreviation",how="outer").drop("abbreviation",1)
merged.head()
```

```python
merged.isnull().any()
```

```python
merged[merged["population"].isnull()].head()
```

```python
merged[merged["state"].isnull()].head()
```

```python
merged.loc[merged["state"].isnull(),"state/region"].unique()
```

```python
merged.loc[merged["state/region"]=="PR","state"]="Puerto Rico"
```

```python
merged.loc[merged["state/region"]=="USA","state"]="Unite states"
```

```python
merged.isnull().any()
```

```python
final=pd.merge(merged,area,on="state",how="left")
final.head()
```

```python
final.isnull().any()
```

```python
final.loc[final["area (sq. mi)"].isnull(),"state"].unique()
```

```python
final.dropna(inplace=True)
```

```python
final.isnull().any()
```

```python
data2010=final.loc[(final["year"]==2010) & (final["ages"]=="total")]
data2010.head()
```

```python
data2010.set_index("state",verify_integrity=True,inplace=True)
data2010.head()
```

```python
data2010["density"]=data2010["population"]/data2010["area (sq. mi)"]
data2010.head()
```

```python
data2010.sort_values("density",ascending=False,inplace=True)
data2010.head()
```
